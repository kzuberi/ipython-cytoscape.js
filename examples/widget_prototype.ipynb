In version 2.0 IPython introduced machinery supporting [javascript widgets](http://ipython.org/ipython-doc/2/whatsnew/version2.0.html#interactive-widgets) that can be synchronized with notebook data. The implementation uses the bootstrap framework. 

Let's test it out by connecting a cytoscape.js graph to tabular data in a pandas dataframe. 

```python
from IPython.html import widgets
from IPython.display import display, Javascript
from IPython.utils.traitlets import Unicode, List
import json
import pandas as pd
```

We'll use a pair of DataFrames, one for nodes and one for edges, and follow the naming conventions used by cytoscape.js to control weight, color, shape, etc.

```python
nodes = pd.DataFrame([['a', 'alpha', 10, 'red', 'hexagon'], 
                      ['b', 'beta', 50, 'blue', 'circle'], 
                      ['g', 'gamma', 80, 'green', 'square']], 
                     columns=['id', 'name', 'weight', 'color', 'shape'])

edges = pd.DataFrame([['a', 'b', 50, 'yellow' ], 
                      ['a', 'g', 20, 'grey'], 
                      ['b', 'g', 70, 'orange']], 
                     columns=['source', 'target', 'strength', 'color'])
display(nodes)
display(edges)
```

Setup the [graph style](http://cytoscape.github.io/cytoscape.js/#style) in a python dict mapping the selector directly to the css.  We've set things up so that we can use the colors and shapes specified in the node/edge DataFrames.

```python
default_style = {'node': {
        'shape': 'data(shape)',
        'width': 'mapData(weight, 40, 80, 20, 60)',
        'content': 'data(name)',
        'text-valign': 'center',
        'text-outline-width': 2,
        'text-outline-color': 'data(color)',
        'background-color': 'data(color)',
        'color': '#fff',
    },

    ':selected': {
        'border-width': 3,
        'border-color': '#333',
    },

    'edge': {
        'width': 'mapData(strength, 70, 100, 2, 6)',
        'target-arrow-shape': 'triangle',
        'source-arrow-shape': 'circle',
        'line-color': 'data(color)',
        'source-arrow-color': 'data(color)',
        'target-arrow-color': 'data(color)',
    },

    'edge.questionable': {
        'line-style': 'dotted',
        'target-arrow-shape': 'diamond',
    },

    '.faded': {
        'opacity': 0.25,
        'text-opacity': 0
    }
}

def to_cystyle(style):
    '''
    cytoscape likes a list of objects with a pair of keys,
    'selector' and 'css'.
    '''

    stylish_list = []
    for selector, css in style.items():
        stylish_list.append({'selector': selector, 'css': css})

    return json.dumps(stylish_list)


default_style_json = to_cystyle(default_style)
```

```python
class CytoWidget(widgets.DOMWidget):
    _view_name = Unicode('CytoWidgetView', sync=True)
    elements = List(Unicode, sync=True)
    style = Unicode(default_style_json, sync=True)


class CytoDataFrame(object):
    def __init__(self, nodes_df, edges_df):
        self._nodes_df = nodes_df
        self._edges_df = edges_df
        self._widget = CytoWidget()
        self._widget.on_displayed(self._on_displayed)

    def _on_displayed(self, e):
        elements = self._get_elements()
        self._widget.elements = elements

    def show(self):
        display(self._widget)

    def _get_elements(self):
        node_data = self._nodes_df.to_json(orient='records')
        edge_data = self._edges_df.to_json(orient='records')
        elements = [node_data, edge_data]
        return elements

    def show(self):
        display(self._widget)
```

```python
%%javascript

require.config({
    paths: {
        cytoscape: "//cytoscape.github.io/cytoscape.js/api/cytoscape.js-latest/cytoscape.min",
    }
});

   
require(["widgets/js/widget", "cytoscape"], function(WidgetManager, cytoscape){

    var CytoWidgetView = IPython.DOMWidgetView.extend({
        render: function(){
            this.$el.css('height', '500px');
            this.$el.css('width', '500px');
            this.$el.css('position', 'relative');

            var that = this;

            this.$el.cytoscape({
                layout: {
                    name: 'circle',
                },

                zoomingEnabled: false,
                userZoomingEnabled: false,
                pan: { x: 0, y: 0 },
                panningEnabled: false,
                
                ready: function(evt){
                    // clear cytoscape container position cache
                    // to work around scrolling bug
                    var cy = this;
                    that.$el.closest("#notebook").scroll(function(){
                        cy._private.renderer.containerBB = null;
                    });
                },                
            })
            
            return this;
        },

        update: function() {
            var cy = this.$el.cytoscape('get');

            var style = this.model.get('style');
            cy.style().resetToDefault().fromJson($.parseJSON(style));

            var elements = this.model.get('elements');
            var node_data = $.parseJSON(elements[0]);
            var edge_data = $.parseJSON(elements[1]);

            var nodes = [];
            for (var i=0; i<node_data.length; i++) {
                nodes.push({data: node_data[i]});
            }

            var edges = [];
            for (var i=0; i<edge_data.length; i++) {
                edges.push({data: edge_data[i]});
            }

            var elements = {nodes: nodes, edges: edges};
            cy.load(elements);

            return CytoWidgetView.__super__.update.apply(this);
        },

    });

    // Register with widget manager.
    WidgetManager.register_widget_view('CytoWidgetView', CytoWidgetView);
});
```

```python
graph = CytoDataFrame(nodes, edges)
graph.show()
```
